### Visualization (10 points)

Let's continue our journey into learning about visualization. 

One of the main goals now is to learn to take control of your visualizations. Since most people start with <code>matplotlib</code>, let's explore this a bit.

You may not be that familiar with coding to make your plots. Perhaps you have always used the "canned" plots in something like Excel. Nothing wrong with using tools already ready for you to use; however, in data science in particular, we often need to take control over what we are doing because of the extremely diverse set of issues that arise. 

First, let's explore just a very few things that are possible, and give you some code to borrow from later in the course. Your first task is to read this code, comment it and make sure you understand how each plot was made so that you can do it yourself next time. 


```python
import numpy as np #import numpy, matplotlib, and cycler libraries
import matplotlib.pyplot as plt
from matplotlib import style
from cycler import cycler

# Generate sample data
x = np.linspace(0, 10, 100) #creates a sample data from range 1 to 10 of 100 values. 
y1 = np.sin(x) #applies the sin function on each value of the x data that was created and save as y1
y2 = np.cos(x) #similarly the case with cos and assign it to y2
y3 = np.tan(x) #similarly the case with tan and assign it to y3

# Create a 3x3 grid of subplots with different styles
fig, axs = plt.subplots(3, 3, figsize=(18, 18)) #create subplots of 3*3, meaning generate a grid pattern to plot the functions, each having 18*18 size
fig.suptitle('Matplotlib Style and Customization Demo', fontsize=20) #give the title to the plot

# Function to set common properties for all subplots
def set_common_properties(ax, title): #create a function such that the plots will take a default set of properties like actomatically creating a legend, xLabel and ylabel definition
    ax.set_title(title)
    ax.set_xlabel('x') #the x axis is labeled as x
    ax.set_ylabel('y') #the x axis is labeled as y
    ax.legend()

# Plot 1: Default style
axs[0, 0].plot(x, y1, label='sin(x)') #in the first plot(0,0), plot the sin curve and label it as sin(x)
axs[0, 0].plot(x, y2, label='cos(x)') # in the same subplot, plot cos curve
set_common_properties(axs[0, 0], 'Default Style') #the default properties will be assigned for the subplot
axs[0, 0].grid(True, linestyle='--', alpha=0.7) #linestyle tells if the line has to be dotted or continuous, alpha talks about the transparency. 

# Plot 2: Seaborn style
with plt.style.context('fivethirtyeight'): # a very similar graph compared to the previous plot but the style is changed to 538 and 538 is passed as a parameter to set_common_parameters() 
    axs[0, 1].plot(x, y1, label='sin(x)')
    axs[0, 1].plot(x, y2, label='cos(x)')
    set_common_properties(axs[0, 1], '538') 

# Plot 3: ggplot style
with plt.style.context('ggplot'): #djusts the style to emulate ggplot (a popular plotting package for R).
    axs[0, 2].plot(x, y1, label='sin(x)')
    axs[0, 2].plot(x, y2, label='cos(x)')
    set_common_properties(axs[0, 2], 'ggplot Style')

# Plot 4: Custom colors and markers (2nd row)
axs[1, 0].plot(x, y1, color='#FF5733', marker='o', markersize=4, label='sin(x)') #provide the color code, maker, so that the line takes the details of what is provided
axs[1, 0].plot(x, y2, color='#33FF57', marker='s', markersize=4, label='cos(x)') #do the same for cos function as well in the same subplot
set_common_properties(axs[1, 0], 'Custom Colors and Markers')
axs[1, 0].set_facecolor('#F0F0F0') #set a facecolor value

# Plot 5: Filled areas
axs[1, 1].fill_between(x, y1, alpha=0.3, label='sin(x) area') #the area in between the curve and the x-axis is covered by a distinct color. This can be used to get the area under the curve
axs[1, 1].fill_between(x, y2, alpha=0.3, label='cos(x) area')
set_common_properties(axs[1, 1], 'Filled Areas')

# Plot 6: Colorblind-friendly palette
with plt.style.context('seaborn-v0_8-colorblind'):
    # Use the 'colorblind' color cycle from seaborn
    plt.rcParams['axes.prop_cycle'] = plt.cycler(color=plt.cm.Set2.colors) # saves the parameter configurations, sets the color cycle for the plot lines from Set2.
    
    for y, label in zip([y1, y2, y3], ['sin(x)', 'cos(x)', 'tan(x)']): #plots all functions in the same subplot
        axs[1, 2].plot(x, y, label=label)
    set_common_properties(axs[1, 2], 'Colorblind-friendly (Seaborn)')

# Plot 7: Log scale
axs[2, 0].semilogy(x, np.abs(y1), label='|sin(x)|') #semilogy is a matplotlib method which returns the semilogarithmic plot where y axis is on a logarithmic scale.
axs[2, 0].semilogy(x, np.abs(y2), label='|cos(x)|')
set_common_properties(axs[2, 0], 'Log Scale')

# Plot 8: Horizontal bar chart
languages = ['Python', 'Java', 'C++', 'JavaScript', 'R'] #categories that we have considered for bar chart
popularity = [68, 45, 35, 62, 25] #give the frequency of each category
axs[2, 1].barh(languages, popularity) #plot the bar
axs[2, 1].set_title('Horizontal Bar Chart') #set the details like the title, x-axis and y-axis
axs[2, 1].set_xlabel('Popularity')
axs[2, 1].set_ylabel('Programming Language')

# Plot 9: Scatter plot with size and color mapping
n = 50 #give the number of samples we have considered
x = np.random.rand(n) #random values for x and y
y = np.random.rand(n)
colors = np.random.rand(n)                                                                                
sizes = 1000 * np.random.rand(n)

scatter = axs[2, 2].scatter(x, y, c=colors, s=sizes, alpha=0.5, cmap='viridis') #in the last subplot, plot the scatter plot of the x against y values v got
axs[2, 2].set_title('Scatter with Size and Color Mapping') #set the name for the plot, the x axis and the y axis
axs[2, 2].set_xlabel('X values')
axs[2, 2].set_ylabel('Y values')
plt.colorbar(scatter, ax=axs[2, 2], label='Color Value')

# Adjust layout and display the plot
plt.tight_layout()
plt.show()
```

```python
import numpy as np # import the numpy library
import matplotlib.pyplot as plt #import matplotlib.pyplot as plt but which using it, can refer as plt 
from sklearn.datasets import load_iris #load iris dataset
from sklearn.preprocessing import StandardScaler # import scaler library

plt.rcdefaults() # plt will be reset to default settings

iris = load_iris() #load the iris dataset
X = iris.data
feature_names = iris.feature_names # the columns of the iris dataset


scaler = StandardScaler() #create an instance of the standard scaler class

X_scaled = scaler.fit_transform(X) # perform a fit and transform on the iris dataset. Standardization is important because it helps in convergence speed and will help in ML models. 

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6)) #create 2 subplots of size 15*6 each


ax1.scatter(X[:, 0], X[:, 1], alpha=0.5) #on fist plot, plot first col on x axis and 2nd col on y axis
ax1.set_title('Original Data') #set title and labels 
ax1.set_xlabel(feature_names[0])
ax1.set_ylabel(feature_names[1])

ax2.scatter(X_scaled[:, 0], X_scaled[:, 1], alpha=0.5) #on second plot, plot first col on x axis and 2nd col on y axis of the scaled data
ax2.set_title('Scaled Data')
ax2.set_xlabel(f'Scaled {feature_names[0]}') #set title and labels 
ax2.set_ylabel(f'Scaled {feature_names[1]}')

plt.tight_layout() #this function automatically squeezes the subplots into the plot 


fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6)) #creates a second set of subplots

ax1.scatter(X[:, 2], X[:, 3], alpha=0.5) #plot a scatter plot on 3rd and 4th columns
ax1.set_title('Original Data')
ax1.set_xlabel(feature_names[2]) #set the names of the features
ax1.set_ylabel(feature_names[3])

ax2.scatter(X_scaled[:, 2], X_scaled[:, 3], alpha=0.5) #plot a scatter plot on the scaled data for the 3rd and 4th question.
ax2.set_title('Scaled Data') #set title and give the feature names
ax2.set_xlabel(f'Scaled {feature_names[2]}')
ax2.set_ylabel(f'Scaled {feature_names[3]}')

plt.tight_layout()


print("Original data:")
print("Means:", np.mean(X, axis=0))
print("Standard deviations:", np.std(X, axis=0)) #here we give the original data mean and the standard deviations

print("\nScaled data:")
print("Means:", np.mean(X_scaled, axis=0))
print("Standard deviations:", np.std(X_scaled, axis=0)) #for the scaled data, the standard deviation is always going to be 1 because the mean becomes 0.

X_reconstructed = scaler.inverse_transform(X_scaled)#The purpose of the inverse transform is to undo the effect of a previous transformation, such as scaling, standardization, or normalization

max_diff = np.max(np.abs(X - X_reconstructed)) #what is the difference between the transformed data and the original data?
print(f"\nMaximum difference between original and reconstructed data: {max_diff}")

# Homework Task: Comment on every line of this code, explaining what it does and why it's important.
```

```python
# one-hot encoding
onehot = OneHotEncoder()
encoded = onehot.fit_transform(df[['Pclass']])

# create feature names manually
unique_values = sorted(df['Pclass'].unique())
feature_names = [f'Pclass_{val}' for val in unique_values]

# convert to DataFrame
encoded_df = pd.DataFrame(encoded.toarray(), columns=feature_names)
print(encoded_df.head())
```

Pclass is a column in the dataset that contains the values of 1,2, and 3 which denotes the class of the passenger. To encode this, the Onehot encoder is used which creates different rows for each unique value of the column, so 3 separate variables(1,2,3) get created that flags one variable as 1 if the value is that particular column. Eg if the pclass is 2, the second column Pclass_2 gets a flag 1 and other 2 columns will have values 0.

```python
le = LabelEncoder()
df['Sex_encoded'] = le.fit_transform(df['Sex'])
print(df[['Sex', 'Sex_encoded']].head())
```

The labelEncoder converts the categorical data to numerical values. This is helpful when the machine has to convert categorical values to ordinal values like first second and third to 1,2,3. For non-ordinal categorical data (where the categories have no inherent order), label encoding can be misleading. For example, if "cat" is encoded as 0, "dog" as 1, and "mouse" as 2, the model might incorrectly assume that "dog" is somehow "greater than" or "in between" "cat" and "mouse." In such cases, one-hot encoding is preferred.


```python
oe = OrdinalEncoder()
df['Pclass_ordinal'] = oe.fit_transform(df[['Pclass']])
print(df[['Pclass', 'Pclass_ordinal']].head())
```

Unlike a Label Encoder that assigns arbitrary numerical values to categories, an Ordinal Encoder respects the inherent order of the data. 

```python
def custom_binary_encode(value):
    return format(value, '02b')

df['Pclass_binary'] = df['Pclass'].apply(custom_binary_encode)
print(df[['Pclass', 'Pclass_binary']].head())
```

This is a custom binary encoding function. The function converts a number from its numerical form to binary form and is applied to the column Pclass using apply function. The customer_binary_encode is a user defined function. which returns the binary value of a number input. 

```python
freq_encoding = df.groupby('Pclass').size()/len(df)
df['Pclass_freq'] = df['Pclass'].map(freq_encoding)
print(df[['Pclass', 'Pclass_freq']].head())
```

This function is similar to the previous function but the only difference is that the class gets encoded with the frequency of occurance of the particular class. Example, The percentage of occurance of class 3 is 54.9042% hence, the class 3 will get mapped to 0.549042.

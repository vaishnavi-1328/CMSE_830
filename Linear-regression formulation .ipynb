# Linear Regression Derivation (1D)

## Step 1: Define the linear model
Our 1D linear regression model has the form:

$$y = mx + b$$

Where:
- $y$ is the predicted value
- $x$ is the input feature
- $m$ is the slope (coefficient)
- $b$ is the y-intercept

## Step 2: Define the Mean Squared Error (MSE) loss function
For $n$ data points, the MSE is:

$$\text{MSE} = \frac{1}{n} \sum_{i=1}^n (y_i - \hat{y}_i)^2 = \frac{1}{n} \sum_{i=1}^n (y_i - (mx_i + b))^2$$

Where:
- $y_i$ is the actual value
- $\hat{y}_i$ is the predicted value $(mx_i + b)$
- $n$ is the number of data points

## Step 3: Set up the minimization condition
To find the optimal values for $m$ and $b$, we need to minimize the MSE. We do this by taking partial derivatives of MSE with respect to $m$ and $b$, and setting them to zero:

$$\frac{\partial(\text{MSE})}{\partial m} = 0$$
$$\frac{\partial(\text{MSE})}{\partial b} = 0$$

## Step 4: Solve for $m$ (slope)
$$\frac{\partial(\text{MSE})}{\partial m} = 0$$
$$\frac{\partial}{\partial m} \left[\frac{1}{n} \sum_{i=1}^n (y_i - (mx_i + b))^2\right] = 0$$
$$\frac{1}{n} \sum_{i=1}^n 2(y_i - (mx_i + b))(-x_i) = 0$$
$$\frac{-2}{n} \sum_{i=1}^n (y_i - mx_i - b)x_i = 0$$
$$\frac{-2}{n} \left[\sum_{i=1}^n y_ix_i - m\sum_{i=1}^n x_i^2 - b\sum_{i=1}^n x_i\right] = 0$$

$$\sum_{i=1}^n y_ix_i - m\sum_{i=1}^n x_i^2 - b\sum_{i=1}^n x_i = 0 \quad \text{(Equation 1)}$$

## Step 5: Solve for $b$ (y-intercept)
$$\frac{\partial(\text{MSE})}{\partial b} = 0$$
$$\frac{\partial}{\partial b} \left[\frac{1}{n} \sum_{i=1}^n (y_i - (mx_i + b))^2\right] = 0$$
$$\frac{1}{n} \sum_{i=1}^n 2(y_i - (mx_i + b))(-1) = 0$$
$$\frac{-2}{n} \sum_{i=1}^n (y_i - mx_i - b) = 0$$
$$\frac{-2}{n} \left[\sum_{i=1}^n y_i - m\sum_{i=1}^n x_i - nb\right] = 0$$

$$\sum_{i=1}^n y_i - m\sum_{i=1}^n x_i - nb = 0$$
$$b = \frac{1}{n} \left[\sum_{i=1}^n y_i - m\sum_{i=1}^n x_i\right] \quad \text{(Equation 2)}$$

## Step 6: Solve the system of equations
From Equation 2:
$$b = \frac{\sum_{i=1}^n y_i - m\sum_{i=1}^n x_i}{n}$$

Substitute this into Equation 1:
$$\sum_{i=1}^n y_ix_i - m\sum_{i=1}^n x_i^2 - \left[\frac{\sum_{i=1}^n y_i - m\sum_{i=1}^n x_i}{n}\right]\sum_{i=1}^n x_i = 0$$

Simplify:
$$n\sum_{i=1}^n y_ix_i - nm\sum_{i=1}^n x_i^2 - \sum_{i=1}^n y_i\sum_{i=1}^n x_i + m(\sum_{i=1}^n x_i)^2 = 0$$
$$n\sum_{i=1}^n y_ix_i - \sum_{i=1}^n y_i\sum_{i=1}^n x_i = m\sum_{i=1}^n x_i^2 - m\frac{(\sum_{i=1}^n x_i)^2}{n}$$
$$n\sum_{i=1}^n y_ix_i - \sum_{i=1}^n y_i\sum_{i=1}^n x_i = m\left[n\sum_{i=1}^n x_i^2 - (\sum_{i=1}^n x_i)^2\right]$$

Solve for $m$:
$$m = \frac{n\sum_{i=1}^n y_ix_i - \sum_{i=1}^n y_i\sum_{i=1}^n x_i}{n\sum_{i=1}^n x_i^2 - (\sum_{i=1}^n x_i)^2}$$

## Step 7: Final equations
The final equations for linear regression in 1D are:

$$m = \frac{n\sum_{i=1}^n y_ix_i - \sum_{i=1}^n y_i\sum_{i=1}^n x_i}{n\sum_{i=1}^n x_i^2 - (\sum_{i=1}^n x_i)^2}$$

$$b = \frac{\sum_{i=1}^n y_i - m\sum_{i=1}^n x_i}{n}$$

These equations allow us to calculate the optimal slope $(m)$ and y-intercept $(b)$ for our linear regression model, minimizing the Mean Squared Error.

```python
import numpy as np # import necessary libraries
import matplotlib.pyplot as plt 
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split

# Step 1: Create a polynomial of a given order
def create_polynomial_data(order, n_samples=100):
    np.random.seed(0) 
    X = np.linspace(0, 1, n_samples) #create random numbers between 0 and 1
    y_true = sum([np.random.randn() * (X ** i) for i in range(order + 1)]) #create a polynomial function of order n, x+x^2+x^3+....
    return X, y_true

# Step 2: Add noise to the polynomial
def add_noise(y_true, noise_level=0.1): #add noise to the experimental y values
    noise = np.random.normal(0, noise_level, size=len(y_true)) #the random numbers are normalized and length has to match with y.
    return y_true + noise #change the values of y by ading noise

# Step 3: Fit polynomial and plot
def fit_and_plot(X, y, order):
    poly = PolynomialFeatures(degree=order) #write the code to fit the polynomial function. 
    X_poly = poly.fit_transform(X.reshape(-1, 1)) #fit_transform() transforms X by generating new features (polynomial terms) like x,x^2,x^3,depending on the degree you set.
    #it has to be changed to a 2D array because of polynomial transformation
    model = LinearRegression() #Linearregreesion() can fit curves when polynomial features are introduced in the data.
    model.fit(X_poly, y) #fit the data in the polynoimal function
    y_pred = model.predict(X_poly) # use the model to predict the values using x
    
    # Plot original noisy data and fitted polynomial
    plt.scatter(X, y, label='Noisy Data', color='gray', alpha=0.6) #scatter the points
    plt.plot(X, y_pred, label=f'Polynomial Order {order}', color='red') #plot the function curve
    plt.legend() 
    plt.show()
    return model, poly

# Step 4: Fit on training set and calculate MSE on test set
def train_test_evaluate(X, y, order, test_size=0.5):
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=test_size, random_state=42) #split the data into train and test data
    
    poly = PolynomialFeatures(degree=order) 
    X_train_poly = poly.fit_transform(X_train.reshape(-1, 1)) #fit both the training and test data to a polynomial function of degree n
    X_test_poly = poly.transform(X_test.reshape(-1, 1))
    
    model = LinearRegression() 
    model.fit(X_train_poly, y_train) # repeat the same steps as fit and plot to the train and test data
    y_train_pred = model.predict(X_train_poly)
    y_test_pred = model.predict(X_test_poly)
    
    # Calculate MSE on both train and test sets
    mse_train = mean_squared_error(y_train, y_train_pred) #calculate the train mean squared error and the test mean squared error.
    mse_test = mean_squared_error(y_test, y_test_pred) 
    
    return mse_train, mse_test #return the error rates. 

# Step 5: Vary polynomial order and plot MSE vs order
def bias_variance_tradeoff(X, y, max_order):
    mse_train_list, mse_test_list = [], []
    orders = range(2, max_order + 1)
    
    for order in orders:
        mse_train, mse_test = train_test_evaluate(X, y, order)
        mse_train_list.append(mse_train)
        mse_test_list.append(mse_test)
    
    # Plot MSE vs polynomial order
    plt.plot(orders, mse_train_list, label='Train MSE', marker='o') #plot train error 
    plt.plot(orders, mse_test_list, label='Test MSE', marker='o') #plot test error 
    plt.xlabel('Polynomial Order')
    plt.ylabel('Mean Squared Error')
    plt.title('Bias-Variance Tradeoff')
    plt.legend()
    plt.show()

# Step 6: Putting it all together
if __name__ == "__main__":
    # Generate dataset
    order_true = 4  # True polynomial order
    X, y_true = create_polynomial_data(order_true) #create a sample dataset, apply a polynomial function to get the y values
    y_noisy = add_noise(y_true, noise_level=0.7) #add noise to the y values
    
    # Plot the polynomial fit for the true model and noisy data
    fit_and_plot(X, y_noisy, order_true) # plot the data and the polynomial function curve
    
    # Analyze the bias-variance tradeoff
    max_order = 10  # Vary the polynomial degree up to 10
    bias_variance_tradeoff(X, y_noisy, max_order) 
```

```python
import numpy as np # import necessary libraries
import matplotlib.pyplot as plt 
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split

# Step 1: Create a polynomial of a given order
def create_polynomial_data(order, n_samples=100):
    np.random.seed(0) 
    X = np.linspace(0, 1, n_samples) #create random numbers between 0 and 1
    y_true = sum([np.random.randn() * (X ** i) for i in range(order + 1)]) #create a polynomial function of order n, x+x^2+x^3+....
    return X, y_true

# Step 2: Add noise to the polynomial
def add_noise(y_true, noise_level=4): #add noise to the experimental y values
    noise = np.random.normal(0, noise_level, size=len(y_true)) #the random numbers are normalized and length has to match with y.
    return y_true + noise #change the values of y by ading noise

# Step 3: Fit polynomial and plot
def fit_and_plot(X, y, order):
    poly = PolynomialFeatures(degree=order) #write the code to fit the polynomial function. 
    X_poly = poly.fit_transform(X.reshape(-1, 1)) #fit_transform() transforms X by generating new features (polynomial terms) like x,x^2,x^3,depending on the degree you set.
    #it has to be changed to a 2D array because of polynomial transformation
    model = LinearRegression() #Linearregreesion() can fit curves when polynomial features are introduced in the data.
    model.fit(X_poly, y) #fit the data in the polynoimal function
    y_pred = model.predict(X_poly) # use the model to predict the values using x
    
    # Plot original noisy data and fitted polynomial
    plt.scatter(X, y, label='Noisy Data', color='gray', alpha=0.6) #scatter the points
    plt.plot(X, y_pred, label=f'Polynomial Order {order}', color='red') #plot the function curve
    plt.legend() 
    plt.show()
    return model, poly

# Step 4: Fit on training set and calculate MSE on test set
def train_test_evaluate(X, y, order, test_size=0.5):
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=test_size, random_state=42) #split the data into train and test data
    
    poly = PolynomialFeatures(degree=order) 
    X_train_poly = poly.fit_transform(X_train.reshape(-1, 1)) #fit both the training and test data to a polynomial function of degree n
    X_test_poly = poly.transform(X_test.reshape(-1, 1))
    
    model = LinearRegression() 
    model.fit(X_train_poly, y_train) # repeat the same steps as fit and plot to the train and test data
    y_train_pred = model.predict(X_train_poly)
    y_test_pred = model.predict(X_test_poly)
    
    # Calculate MSE on both train and test sets
    mse_train = mean_squared_error(y_train, y_train_pred) #calculate the train mean squared error and the test mean squared error.
    mse_test = mean_squared_error(y_test, y_test_pred) 
    
    return mse_train, mse_test #return the error rates. 

# Step 5: Vary polynomial order and plot MSE vs order
def bias_variance_tradeoff(X, y, max_order):
    mse_train_list, mse_test_list = [], []
    orders = range(2, max_order + 1)
    
    for order in orders:
        mse_train, mse_test = train_test_evaluate(X, y, order)
        mse_train_list.append(mse_train)
        mse_test_list.append(mse_test)
    
    # Plot MSE vs polynomial order
    plt.plot(orders, mse_train_list, label='Train MSE', marker='o') #plot train error 
    plt.plot(orders, mse_test_list, label='Test MSE', marker='o') #plot test error 
    plt.xlabel('Polynomial Order')
    plt.ylabel('Mean Squared Error')
    plt.title('Bias-Variance Tradeoff')
    plt.legend()
    plt.show()

# Step 6: Putting it all together
if __name__ == "__main__":
    # Generate dataset
    order_true = 4  # True polynomial order
    X, y_true = create_polynomial_data(order_true) #create a sample dataset, apply a polynomial function to get the y values
    y_noisy = add_noise(y_true, noise_level=1) #add noise to the y values
    
    # Plot the polynomial fit for the true model and noisy data
    fit_and_plot(X, y_noisy, order_true) # plot the data and the polynomial function curve
    
    # Analyze the bias-variance tradeoff
    max_order = 20  # Vary the polynomial degree up to 12
    bias_variance_tradeoff(X, y_noisy, max_order) 
```

- Data generation: Creates a polynomial dataset of a specified order and adds noise to simulate real-world data.
- Polynomial fitting: Uses scikit-learn to fit polynomials of various degrees to the noisy data.
- Visualization: Plots the original data and the fitted polynomial curves.
- Train-test split: Divides the data into training and testing sets to evaluate model performance.
- Error calculation: Computes Mean Squared Error (MSE) for both training and testing data.
- Bias-variance analysis: Varies the polynomial order and plots the training and testing MSE to illustrate the bias-variance tradeoff.
